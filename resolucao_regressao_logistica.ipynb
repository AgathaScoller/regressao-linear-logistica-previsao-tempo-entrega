# Resolução Completa – Regressão Linear com Base de Logística

Objetivo: prever o tempo de entrega (`tempo_entrega_min`).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# ajuste o caminho do arquivo
df = pd.read_csv('sua_base.csv')
df.head()
```

# Parte E – Regressão Simples

```python
X = df[['distancia_km']]
y = df['tempo_entrega_min']

modelo = LinearRegression()
modelo.fit(X, y)

coef = modelo.coef_[0]
intercepto = modelo.intercept_

y_pred = modelo.predict(X)
```

```python
resultado = pd.DataFrame({
    'real': y,
    'previsto': y_pred,
    'erro': y - y_pred
})
resultado.head()
```

```python
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

mae, rmse, r2
```

# Parte F – Reta

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

# Parte H – Dummies

```python
df_modelo = pd.get_dummies(df, columns=['trafego','clima','tipo_entrega','regiao'], drop_first=True)
df_modelo.head()
```

# Parte I – Regressão Múltipla

```python
X = df_modelo.drop(columns=['id_pedido','tempo_entrega_min','atrasou'])
y = df_modelo['tempo_entrega_min']

modelo = LinearRegression()
modelo.fit(X, y)

y_pred = modelo.predict(X)

mean_absolute_error(y, y_pred), np.sqrt(mean_squared_error(y, y_pred)), r2_score(y, y_pred)
```

# Parte K – Treino/Teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)

mean_absolute_error(y_test, y_pred), np.sqrt(mean_squared_error(y_test, y_pred)), r2_score(y_test, y_pred)
```

# Conclusão

- O modelo múltiplo apresenta melhor desempenho
- Variáveis mais importantes: distância, tráfego e clima
- O modelo é útil, mas não deve ser usado sozinho
- Melhorias: dados em tempo real e modelos mais avançados
